```python
'''
Simple example showing how to:
- create a .csv file with custom column headers and row numbers
- view .csv file in its entirety, or just the first 5 entries of it (pd.head())
- use numpy version of head()
- use matplotlib to visualise data
'''

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## creating csv

```python
import csv
# Create list of data 
data = []
for i in range(1, 51):
    data.append([i, i + 50, i + 100])

file_name = "testdata.csv"

# Write the data to a CSV file
with open(file_name, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(['apple', 'banana', 'orange'])
    writer.writerows(data)

# REMEMBER: easy way to print log output of string with your variable is using print(f'blah blah {VARIABLE NAME} blah blah')
print(f"'{file_name}' created")
```

## reading csv with pandas

```python
df = pd.read_csv(file_name)
print(df)
#print first five rows with 'head'. works only with pandas' dataframes.
print(df.head())
```

## data ranges

```python
#crop a range you want
minrange = 4
maxrange = 8
df_cropped = df.iloc[minrange:maxrange]
print(df_cropped)
#for tail-end, leave end empty
startofend = 44
df_cropped_tail = df.iloc[startofend:]
print(df_cropped_tail)
```

```python
#select variable you wanna focus on
variable = 'banana'
minrange = 4
maxrange = 8
df_cropped_singlevar = df[variable].iloc[minrange:maxrange]
print(df_cropped_singlevar)
```

## numpy version

```python
#in numpy, there's no head(). instead:
df_numpy = df.to_numpy()
print(df_numpy[:5])
#for tail end, similarly:
print(df_numpy[44:])
#for variable focusing, deal w it using indexing
print(df_numpy[:5, 1])
```

## matplotlib visualisation

```python
#to plot multiple plots in a single plot, use subplot.
plt.subplot(3,2,1) #format: 3 by 2 plot, and this subplot is number 1 (goes top to bot, then left to right)
plt.plot(df['apple'], color = 'r', label='applelabel')#can specify color, legend
plt.plot(df['banana'], color = 'y', label='bananalabel')
plt.title('fruits and their numbers')
plt.legend(bbox_to_anchor=(0,0))#tell matplotlib to show legend. 
                                    #can also specify where to show legend.top right=1,1
# plt.show() #show plot 

plt.subplot(3,2,2)#second subplot (on the right now)
t = np.arange(50,100,1) #u can set ur own range to act as one axis. 
                        #here i create t, ascending range from 10 to 100, to act as my x axis.
                        #note that if u have 50 entries, both x and y axis must be a range of 50 as well.
                        #note the x axis range here?
plt.plot(t, df['apple'], color='r', label='applelabel') #xaxis by yaxis)
plt.plot(t, df['banana'], color='y', label='bananalabel') #xaxis by yaxis)
plt.title('fruits and their numbers part II')
plt.legend(bbox_to_anchor=(0.5,0.5))
# plt.show()

plt.subplot(3,2,4) #third subplot. 
t = np.arange(50,100,1) 
plt.plot(t, df['apple'], color='r', label='applelabel') #xaxis by yaxis)
plt.plot(t, df['banana'], color='y', label='bananalabel') #xaxis by yaxis)
plt.plot(t, df['orange'], color='b', label='orangelabel') #xaxis by yaxis)
plt.title('fruits and their numbers part III')
plt.legend(bbox_to_anchor=(1,0))


plt.subplots_adjust( right = 2, top = 1.5) #for finetuning. if i use right=2, im stretching the subplots
                                #to 2 times the original right-side ending pt.
plt.show()
```
